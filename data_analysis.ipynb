```python
# tdf['Impact'].value_counts()
# tdf['Priority'].value_counts()
# tdf['Status'].value_counts()
# tdf['Type'].value_counts()
tdf['Item'].value_counts()
print(df['Item'].value_counts())
```

```python
tdf['Impact']=tdf['Impact'].map({"Low":-1,"Medium":0,"High":1})
tdf['Priority']=tdf['Priority'].map({"Low":-1,"Medium":0,"High":1,"Urgent":2})
tdf['Status']=tdf['Status'].map({"Closed":1})
tdf['Type']=tdf['Type'].map({"Service Request":1,"Incident":0})
tdf['Item']=tdf['Item'].map({"False Positive":-1,"Benign":0,"True Positive":1,"Report":2,"Wireless":3})
```

```python
tdf
```

```python
tdf['Category']=le.fit_transform(tdf['Category'])
print(le.classes_)
mapping = dict(zip(le.classes_, le.transform(le.classes_)))
mapping
```

```python
tdf['Sub-Category']=le.fit_transform(tdf['Sub-Category'])
mapping = dict(zip(le.classes_, le.transform(le.classes_)))
mapping
```

```python
X=tdf[['Impact','Priority','Status','Type','Category','Sub-Category','Resolution Time']]
y=tdf['Item']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=42,test_size=0.3)
dtc=DecisionTreeClassifier()
dtc.fit(X_train,y_train)
```

```python
pred=dtc.predict([[0,0,1,0,6,1,215.4333]])
y_pred=dtc.predict(X_test)
pred
```

```python
from sklearn.metrics import accuracy_score, classification_report
print(accuracy_score(y_test, y_pred)*100)
print(classification_report(y_test, y_pred))
```
